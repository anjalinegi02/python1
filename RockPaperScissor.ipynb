```python
import tkinter as tk
from tkinter import messagebox

# Initialize variables for player choices
player1_choice = None
player2_choice = None

# Function to determine the winner
def determine_winner():
    global player1_choice, player2_choice
    
    if player1_choice == player2_choice:
        result = "It's a Tie!"
    elif (player1_choice == "Rock" and player2_choice == "Scissors") or \
         (player1_choice == "Paper" and player2_choice == "Rock") or \
         (player1_choice == "Scissors" and player2_choice == "Paper"):
        result = "Player 1 Wins!"
    else:
        result = "Player 2 Wins!"
    
    # Show result and reset
    messagebox.showinfo("Result", f"Player 1 chose: {player1_choice}\nPlayer 2 chose: {player2_choice}\n\n{result}")
    reset_game()

# Function to handle player choices
def player_choice(player, choice):
    global player1_choice, player2_choice
    
    if player == 1:
        player1_choice = choice
        info_label.config(text="Player 2: Make your choice!", fg="blue")
    elif player == 2:
        player2_choice = choice
        determine_winner()

# Function to reset the game
def reset_game():
    global player1_choice, player2_choice
    player1_choice = None
    player2_choice = None
    info_label.config(text="Player 1: Make your choice!", fg="green")

# Create the main window
root = tk.Tk()
root.title("Rock Paper Scissors - Multiplayer")
root.geometry("400x400")
root.config(bg="lavender") 

# Information label
info_label = tk.Label(root, text="Player 1: Make your choice!", font=("Arial", 16, "bold"), fg="green", bg="lavender")
info_label.pack(pady=20)

# Button styling
button_style = {
    "width": 15,
    "height": 2,
    "font": ("Arial", 12, "bold"),
    "relief": "raised",
    "bd": 4,
}

# Rock button
rock_button = tk.Button(root, text="Rock", bg="pink", fg="black", **button_style,  
                        command=lambda: player_choice(1 if player1_choice is None else 2, "Rock"))
rock_button.pack(pady=10)

# Paper button
paper_button = tk.Button(root, text="Paper", bg="light blue", fg="black", **button_style,  
                         command=lambda: player_choice(1 if player1_choice is None else 2, "Paper"))
paper_button.pack(pady=10)

# Scissors button
scissors_button = tk.Button(root, text="Scissors", bg="plum", fg="black", **button_style,  
                            command=lambda: player_choice(1 if player1_choice is None else 2, "Scissors"))
scissors_button.pack(pady=10)

# Footer
footer_label = tk.Label(root, text="Enjoy the Game!", font=("Arial", 10, "italic"), bg="lavender", fg="gray")
footer_label.pack(side="bottom", pady=10)

# Run the GUI event loop
root.mainloop()
```
